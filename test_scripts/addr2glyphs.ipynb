This notebook is for developing a small tool to read all address images from a folder, split those into glyph images, and dump them in another folder.

```python
import os
from os import listdir
from PIL import Image
import re
```

```python
def isolate_glyph(img, ind):
	'''Isolates one of 12 glyphs in a 768x64 addr image
	'''
	# Setting the points for cropped image
	left = 0 + ind*64
	right = left+64
	top = 0
	bottom = 64
	
	# Cropped image of above dimension
	# (It will not change original image)
	return img.crop((left, top, right, bottom))
```

```python
# get the path/directory
addr_dir = "C:\\csci490\\images\\addr"
out_path = "C:\\csci490\\images\\glyph\\unlabeled"
addresses = os.listdir(addr_dir)
```

```python
interval = 25
start = 110
end = start + interval
```

```python

# for fname in addresses:
for i in range(210, len(addresses)):
    # check if the fname ends with png
    fname = addresses[i]
    path = addr_dir + "\\" + fname
    if (fname.endswith(".png")):
        addr = Image.open(path)
        # print(paths)
        for index in range(12):
            glyph = isolate_glyph(addr, index)
            dest = out_path + "\\" + fname[:-4] + "_%02d.png"%index 
            glyph.save(dest)

# start = end
# end += interval
```
