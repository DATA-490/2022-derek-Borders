```python
def isolate_glyph(img, ind):
	'''Isolates 12 glyphs and stores them in an array
	'''
	# Setting the points for cropped image
	left = 23 + ind*64
	right = 87 + ind*64
	top = height - 135
	bottom = height - 65 
	
	# Cropped image of above dimension
	# (It will not change original image)
	return img.crop((left, top, right, bottom))
```

```python
# Isolate each glyph
glyphs = [isolate_glyph(im, i) for i in range(12)]

# [print(i) for i in range(12)]
# glyphs = []
# for i in range(12):
# 	glyphs.append(isolate_glyph(im, i))  

for x in range(12):
	glyphs[x].show() 

# type(glyphs[0])
```
